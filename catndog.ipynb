```python
pic=r"C:\Users\sobee\Desktop\PetImages"
categories=["Cat","Dog"]
```

Dosya local yolla indirildi ve notebook içine yüklendi

```python
img_size=32
imgs=[]
```

Fotoğraflar için ortak bir 'size' belirlendi ve data adlı liste oluşturuldu. Size değeri ise durumunda değiştirebilir [0,300] arasında bir random değer atanarak tekrardan işlem yaptırılabilir.

```python
for category in categories:
    fl=os.path.join(pic,category)
    lbl=categories.index(category)
    for img in os.listdir(fl):
        try:
            img_path=os.path.join(fl,img)
            img_arr=cv2.imread(img_path)
            print("Done")
            img_arr=cv2.resize(img_arr,(img_size,img_size))
            imgs.append([img_arr,lbl])
        except Exception as e:
            print(e)
```

Notebook içine yüklediğimiz görselleri ve category kısmını birleştirdikten sonra bu görselleri bir array haline getirip aynı boyutta olmaları için boyutları aynı olacak şekilde değiştirilip ortak bir listeye(imgs) eklenir. Try-Except kullanma nedenimiz döngü içine girerken dosya içindeki bazı görselleri cv2.imread okuyamadığı için döngüyü hataya zorluyor. cv2.imread'in okuduğu değerlerle devam ediyoruz.

```python
x=[]
y=[]

for features,labels in imgs:
  x.append(features)
  y.append(labels)
```

```python
type(x)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
print("x_train shape: ",x_train.shape)
print("y_train shape: ",y_train.shape)
print("x_test shape: ",x_test.shape)
print("y_test shape: ",y_test.shape)
```
